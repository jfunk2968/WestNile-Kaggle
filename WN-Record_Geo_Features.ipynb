```python
%matplotlib inline

import pandas as pd
import numpy as np
import funktions as fk
from datetime import datetime
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
bins1 = np.linspace(min(test['Latitude']),max(test['Latitude']), 10)
test['lat'] = np.digitize(test['Latitude'], bins1)
train['lat'] = np.digitize(train['Latitude'], bins1)

bins2 = np.linspace(min(test['Longitude']),max(test['Longitude']), 10)
test['lng'] = np.digitize(test['Longitude'], bins2)
train['lng'] = np.digitize(train['Longitude'], bins2)


print test['lat'].value_counts(dropna=False)
print train['lat'].value_counts(dropna=False)
print test['lng'].value_counts(dropna=False)
print train['lng'].value_counts(dropna=False)
```

```python
lat_test = test['lat'].astype(str)
lng_test = test['lng'].astype(str)
test['geo'] = pd.DataFrame([lat_test,lng_test]).apply(lambda x: '-'.join(x))

lat_train = train['lat'].astype(str)
lng_train = train['lng'].astype(str)
train['geo'] = pd.DataFrame([lat_train,lng_train]).apply(lambda x: '-'.join(x))
```

```python
# training set has only records where skeeters caught, so need to pad with filler rows to reflect scoring set
train_cnt = train.groupby(['Date','geo','Species'])['WnvPresent'].count().reset_index()
train_cnt.rename(columns = {'WnvPresent':'grec_cnt'}, inplace = True)
#train_cnt.head()

# all combos of training dates-traps-species (test has 1 additional Species value ...)
import itertools
lists = [list(train['Date'].unique()),list(train['geo'].unique()),list(test['Species'].unique())]
all_comb = pd.DataFrame(list(itertools.product(*lists)),columns=['Date','geo','Species'])
#all_comb.head()

train_cnt2 = all_comb.merge(train_cnt,left_on=['Date','geo','Species'],right_on=['Date','geo','Species'],how='left')
train_cnt2.fillna(1,inplace=True)
#train_cnt2.head(20)

# test set has at least 1 record for every date-trap-species combo
test_cnt = test.groupby(['Date','geo','Species'])['Id'].count().reset_index()
test_cnt.rename(columns = {'Id':'grec_cnt'}, inplace = True)
#test_cnt.head()
```

```python
date_format = "%Y-%m-%d"
train_cnt2['dt'] = train_cnt2['Date'].apply(lambda x: datetime.strptime(x, date_format))
test_cnt['dt']  = test_cnt['Date'].apply(lambda x: datetime.strptime(x, date_format))
```

```python
def agger(df,days,date):
    chunk = df['dt'].apply(lambda x: ((0 <= (date - x).days) & ((date - x).days < days)))
    _mean = np.sum(df['grec_cnt'].loc[chunk])
    _min = min(df['grec_cnt'].loc[chunk])
    _max = min(df['grec_cnt'].loc[chunk])
    return _mean, _min, _max
```

```python
geos = []
for geo in train_cnt2['geo'].unique():
    time = datetime.now()
    for spec in train_cnt2['Species'].unique():
        temp = train_cnt2.loc[((train_cnt2['geo']==geo) & (train_cnt2['Species']==spec))].copy()

        for d in [3,7,14,30,60]:
            
            recs = temp['dt'].apply(lambda x: agger(temp,d,x))  
            
            temp['grec_cnt_'+str(d)+'_sum']= recs.apply(lambda x: x[0])    
            temp['grec_cnt_'+str(d)+'_2']= temp['grec_cnt_'+str(d)+'_sum'].apply(lambda x: 1 if x>1 else 0) 
            temp['grec_cnt_'+str(d)+'_max']= recs.apply(lambda x: x[2])   
            temp['grec_cnt_'+str(d)+'_min']= recs.apply(lambda x: x[1])    

        geos.append(temp)
    print geo,(datetime.now()-time).seconds
```

```python
train_geos_cnt = pd.concat(geos,axis=0)
print len(train_geos_cnt),len(train_cnt2)
```
