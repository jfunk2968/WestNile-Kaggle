```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import funktions as fk
```

## Explore Training Data

```python
train = pd.read_csv('train.csv')
train.head().T
```

```python
train['main_trap'] = train['Trap'].apply(lambda x: x[0:4])
train['year'] = train['Date'].apply(lambda x: x[0:4])
train['month'] = train['Date'].apply(lambda x: x[5:7])
```

```python
# count of records by year-mth
train['Date'].apply(lambda x: x[0:7]).value_counts().sort_index().plot()
```

```python
# how many unique trap locations
(train['Latitude'].apply(lambda x: str(x)) + train['Longitude'].apply(lambda x: str(x))).nunique()
```

```python
# how many unique traps and main traps
train['Trap'].nunique(), train['main_trap'].nunique()   
```

```python
# only 2 satelite traps in data set ... shouldn't be much an issue
train['Trap'].loc[train['Trap'].apply(lambda x: len(x)>4)]
```

```python
# distribution of trap record counts by date/species
train.groupby(['Date','Trap','Species'])['NumMosquitos'].count().value_counts().sort_index().plot(kind='bar')
```

```python
# let's look at some highly repeated records
c = train.groupby(['Date','Trap','Species'])['NumMosquitos'].count()
c.loc[c>30]
```

```python
# view a few records for a sample loc/date/species with many entries
train.loc[((train['Date']=='2007-07-25') & (train['Trap']=='T115') & (train['Species']=='CULEX PIPIENS'))].T
```

```python
# why are there multiple records for this trap and date ...?
train.loc[((train['Date']=='2013-09-26') & (train['Trap']=='T900'))].T
```

```python
# no details provided for this variables, so not sure what to use it for initialy
# could be a useful filter if any geo variables are very location specific ...
train['AddressAccuracy'].value_counts()
```

```python
# record level target rate is low
train['WnvPresent'].value_counts(normalize=True)
```

```python
train['NumMosquitos'].hist(bins=20)
```

```python
# do traps always stay in same location?
train.groupby(['Trap'])['Address'].nunique().value_counts()
# only one trap looks like it moved in the training data
```

## Explore Test Data

```python
test['main_trap'] = test['Trap'].apply(lambda x: x[0:4])
test['year'] = test['Date'].apply(lambda x: x[0:4])
test['month'] = test['Date'].apply(lambda x: x[5:7])
```

```python
list(test.columns)
```

```python
#test.groupby(['Date','Trap','Species'])['Id'].count().value_counts()
test.groupby(['Date','Trap'])['Id'].count().value_counts().sort_index()

# interesting that there are duplicate records, indicated by combos with >8 rows
# no mosquito counts available in the test data
# should probably add leakage features based on record count for kaggle comp
```

```python
test['Date'].apply(lambda x: x[0:4]).value_counts().sort_index()
```

```python
# do train and test have same set of trap locations?
set(test['Address']) ==  set(train['Address']), test['Address'].nunique(), train['Address'].nunique()  
```

```python
# 13 locations in test not included in training
set(test['Address']).difference(set(train['Address']))
```

```python
# all locations in training also included in test
set(train['Address']).difference(set(test['Address']))
```

```python
# do traps always stay in same location?
test.groupby(['Trap'])['Address'].nunique().value_counts()
# again only one trap looks like it moves in test data also
```
